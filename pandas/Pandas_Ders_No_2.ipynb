## <p style="background-color:#002B5B; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">1.1 Series Oluturma (4 yolla oluturulabilir)</p>

## I - Liste veya Tuple Kullanarak Seri Oluturmak

```python
# Bir listeden seri oluturabiliriz
items = (1,2,3,4)
pd.Series(items)
```

## II - Numpy Array Kullanarak Seri Oluturmak

```python
# Numpy array den bir seri oluturabiliriz.
veri  = np.array(['x','y','z'])
pd.Series(veri)
```

## III - Dictionary Kullanarak Seri Oluturmak

```python
# Dictionary den bir seri oluturur.
veri = { 'uk':'united kingdom'
        ,'fr':'france' }
pd.Series(veri)
```

## IV - Scalar le Seri Oluturmak 

<div class="alert alert-info">Scalar deyince aklmza Lineer Cebir deki veya fizik deki scalar deerler gibi dnmeyin. Scalar diyince aklmza list veya tuple gibi yaplarn aksine tek bir birimden oluan yaplar gelmeli. bunlar integer boolean gibi yaplardr</div>

```python
isim = 'Ali'
sc_ser = pd.Series(isim)
sc_ser
```

```python
sc_ser = pd.Series(isim, index=range(3))
sc_ser
```

```python
sc_ser = pd.Series(isim, index=['a','b','c'])
sc_ser
```

 indexi belli olan bir seride farkl bir index belirlersek ne olur?

```python
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = pd.Series(sdata)
obj3
```

```python
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = pd.Series(sdata, index=states)
obj4
```

 sdata bulunan verilerin karlklar geldi. ancak California sdata da bulunmadndan deeri NaN (not a number) olarak grnd. sdata da bulunan Utah, indexde yer almadndan sonu objesinde yer almad.

 Biz bu NaN tipindeki deerleri bundan sonra missing value olarak ifade edeceiz.

 missing value leri tespit etmek iin isnull veya notnull fonksiyonlar kullanabiliriz.

```python
pd.isnull(obj4)
```

```python
pd.notnull(obj4)
```

```python
obj4.isnull()
```
