###  Dictionary ile Dataframe Oluşturma

```python
v1 = np.random.randint(1, 10, size=4)
v2 = np.random.randint(1, 10, size=4)
v3 = np.random.randint(1, 10, size=4)
```

```python
v1,v2,v3
```

```python
dict1 = {'sutun_1':v1, 'sutun_2':v2, 'sutun_3':v3}
dict1
```

```python
df = pd.DataFrame(data= dict1)
df
```

⭐ Dataframe oluşturmada kullanılan dataların diğer bir yaygın biçimi iç içe dictionarylerdir. 

```python
pop = {'Nevada': {2001: 2.4, 2002: 2.9},
       'Ohio': {2000: 1.5, 2001: 1.7, 2002: 3.6}}
pop
```

Bu şekilde bir yapıdan Dataframe oluşturmak istediğimizde pandas dışardaki dictionary nin key değerlerini sütün olarak değerlendirir. İçerideki dictionary key değerlerini de index olarak değerlendirir. 

```python
frame3 = pd.DataFrame(pop)
frame3
```

```python
frame3.T
```

```python
pd.DataFrame(pop, index=[2000, 2001, 2002,2003])
```

Dataframe in indexine ve sütünlarına isim vermek

```python
frame3.index.name = 'year'; frame3.columns.name = 'State'
frame3
```
